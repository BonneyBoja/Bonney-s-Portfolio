```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
df=pd.read_csv('homeprices.csv')
df.head()
```

```python
# Check dataset structure
df.info()
df.describe()
df.isnull().sum()
```

```python
 #Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df['area'],df['price'],color='red',marker='+')
plt.xlabel('Area(sq ft)')
plt.ylabel('Price($)')
plt.title('Area vs Price')
plt.show()
```

```python
# Prepare data
x= df.iloc[:,[0]]# Feature column (area)
y=df.iloc[:,[1]] # Target column (price)
```

```python
#split into training and testing set (80% training and 20% testing)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Build the model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# Predictions
y_pred=model.predict(x_test)
```

```python
# Evaluate model
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)
print("\nModel Evaluation Metrics:")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R Score: {r2}")
```

```python
# Visualize regression line
plt.figure()
plt.scatter(x, y,color='red',marker='+')
plt.plot(x, model.predict(x), linewidth=2,color='red')
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ($)")
plt.title("Regression Line vs Actual Data")
plt.show()
```
